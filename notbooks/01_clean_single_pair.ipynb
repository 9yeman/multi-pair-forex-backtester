```python
import pandas as pd



file_name = "EURUSDM1.csv"

df = pd.read_csv("EURUSDM1.csv", header=None, encoding='utf-16')


print(df.head(10))
print("--------------------information")
print(df.info())
print("--------------------describe")
print(df.describe())
```

```python
df = df.iloc[:, :-1]
```

```python
df.columns = ['DateTime', 'Open', 'High', 'Low', 'Close', 'Volume']

df['DateTime'] = pd.to_datetime(df['DateTime'], format='%Y.%m.%d %H:%M')

df.to_csv("EURUSDM1_clean.csv", index=False)

print(df.head())
```

```python
print(df.tail())
```

```python
num_duplicates = df.duplicated().sum()
print(f"duplicated: {num_duplicates}")
```

```python
print(df.isnull().sum())  
```

```python
df = df.sort_values('DateTime')  
df = df.reset_index(drop=True)
```

```python
timeframes = {
    '1min': '1T',
    '5min': '5T',
    '15min': '15T',
    '30min': '30T',
    '1h': '1H',
    '4h': '4H',
    '1d': '1D'
}
```

```python
for name, rule in timeframes.items():
    df_resampled = df.resample(rule, on='DateTime').agg({
        'Open':'first',
        'High':'max',
        'Low':'min',
        'Close':'last',
        'Volume':'sum'
    })
    df_resampled = df_resampled.dropna()
    df_resampled = df_resampled.reset_index()
    df_resampled.to_csv(f"EURUSDM1_{name}.csv", index=False)
    print(f"time frame saved {name} in EURUSDM1_{name}.csv")
```
